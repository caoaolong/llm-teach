```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# 设置随机种子（保证可复现）
torch.manual_seed(42)


# 生成正弦波数据
def generate_sin_wave(length=1000, periods=5):
    x = np.linspace(0, 2 * np.pi * periods, length)
    y = np.sin(x)
    return y.reshape(-1, 1)  # 转换为 (seq_len, 1)


# 数据预处理
def create_dataset(_data, _look_back=10):
    _X, _Y = [], []
    for i in range(len(_data) - _look_back):
        _X.append(_data[i:i + _look_back])
        _Y.append(_data[i + _look_back])
    return torch.FloatTensor(np.array(_X)), torch.FloatTensor(np.array(_Y))


# 参数设置
look_back = 20  # 用前20个点预测下一个点
hidden_size = 32
epochs = 80
lr = 0.01

# 生成数据并归一化
data = generate_sin_wave()
scaler = MinMaxScaler(feature_range=(-1, 1))
data = scaler.fit_transform(data)
X, Y = create_dataset(data, look_back)

# 划分训练集和测试集
train_size = int(len(X) * 0.8)
X_train, Y_train = X[:train_size], Y[:train_size]
X_test, Y_test = X[train_size:], Y[train_size:]


# 定义RNN模型
class SimpleRNN(nn.Module):
    def __init__(self, _input_size=1, _hidden_size=32, output_size=1):
        super(SimpleRNN, self).__init__()
        self.rnn = nn.RNN(_input_size, _hidden_size, batch_first=True)
        self.fc = nn.Linear(_hidden_size, output_size)

    def forward(self, x):
        out, _ = self.rnn(x)  # out: (batch, seq_len, hidden_size)
        out = self.fc(out[:, -1, :])  # 只取最后一个时间步的输出
        return out


# 初始化模型、损失函数和优化器
model = SimpleRNN(_input_size=1, _hidden_size=hidden_size)
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=lr)

# 训练模型
train_losses = []
test_losses = []

for epoch in range(epochs):
    model.train()
    optimizer.zero_grad()
    outputs = model(X_train)
    loss = criterion(outputs, Y_train)
    loss.backward()
    optimizer.step()
    train_losses.append(loss.item())

    # 测试集损失
    model.eval()
    with torch.no_grad():
        test_outputs = model(X_test)
        test_loss = criterion(test_outputs, Y_test)
        test_losses.append(test_loss.item())

    if (epoch + 1) % 10 == 0:
        print(f'Epoch [{epoch + 1}/{epochs}], Train Loss: {loss.item():.4f}, Test Loss: {test_loss.item():.4f}')

example_input = torch.randn(1, 10, 1)
torch.onnx.export(
    model, example_input, "../server/models/sin.onnx",
    input_names=["text"],
    output_names=["output"],
    dynamic_axes={"text": {0: "seq_len", 1: "batch_size"}, "output": {0: "batch_size"}},
    opset_version=11
)

print("✅ 模型已导出为 sin.onnx")

# 绘制训练和测试损失曲线
plt.figure(figsize=(12, 5))
plt.plot(train_losses, label='Train Loss')
plt.plot(test_losses, label='Test Loss')
plt.title('Training and Testing Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

# 预测并可视化结果
model.eval()
with torch.no_grad():
    train_predict = model(X_train).numpy()
    test_predict = model(X_test).numpy()

# 反归一化
train_predict = scaler.inverse_transform(train_predict)
Y_train = scaler.inverse_transform(Y_train.numpy())
test_predict = scaler.inverse_transform(test_predict)
Y_test = scaler.inverse_transform(Y_test.numpy())

# 绘制真实值和预测值
plt.figure(figsize=(12, 5))
plt.plot(Y_train, label='True Train Values')
plt.plot(train_predict, '--', label='Train Predictions')
plt.plot(range(len(Y_train), len(Y_train) + len(Y_test)), Y_test, label='True Test Values')
plt.plot(range(len(Y_train), len(Y_train) + len(Y_test)), test_predict, '--', label='Test Predictions')
plt.title('Sine Wave Prediction with RNN')
plt.xlabel('Time Step')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from datasets import load_dataset
import matplotlib.pyplot as plt
import re
from collections import Counter
from torch.nn.utils.rnn import pad_sequence

# 设置设备
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")


# ========== 数据加载与预处理 ==========

def simple_tokenizer(text):
    return re.findall(r'\b\w+\b', text.lower())


def build_vocab(_data, tokenizer, min_freq=2):
    counter = Counter()
    for label, text in _data:
        counter.update(tokenizer(text))
    _vocab = {"<pad>": 0, "<unk>": 1}
    for word, freq in counter.items():
        if freq >= min_freq:
            _vocab[word] = len(_vocab)
    return _vocab


def encode(text, _vocab, tokenizer):
    return [_vocab.get(token, _vocab["<unk>"]) for token in tokenizer(text)]


def collate_batch(batch):
    _labels, _texts = zip(*batch)
    encoded = [torch.tensor(encode(text, vocab, simple_tokenizer)) for text in _texts]
    padded = pad_sequence(encoded, padding_value=vocab["<pad>"])
    _labels = torch.tensor(_labels)
    return padded.to(DEVICE), _labels.to(DEVICE)


# 加载 IMDb 数据集
raw_dataset = load_dataset("imdb")
train_data = [(ex["label"], ex["text"]) for ex in raw_dataset["train"]]
test_data = [(ex["label"], ex["text"]) for ex in raw_dataset["test"]]

# 构建词表
vocab = build_vocab(train_data, simple_tokenizer)

# 构造 DataLoader
from torch.utils.data import DataLoader

BATCH_SIZE = 32
train_loader = DataLoader(train_data, batch_size=BATCH_SIZE, shuffle=True, collate_fn=collate_batch)
test_loader = DataLoader(test_data, batch_size=BATCH_SIZE, shuffle=False, collate_fn=collate_batch)


# ========== 模型定义 ==========

class SentimentRNN(nn.Module):
    def __init__(self, vocab_size, embed_dim, hidden_dim, output_dim):
        super().__init__()
        # pytorch 的 Embedding 使用的是 Word2Vec
        self.embedding = nn.Embedding(vocab_size, embed_dim, padding_idx=0)
        self.rnn = nn.LSTM(embed_dim, hidden_dim)
        self.fc = nn.Linear(hidden_dim, output_dim)

    def forward(self, text):  # text: [seq_len, batch_size]
        embedded = self.embedding(text)
        _, (hidden, _) = self.rnn(embedded)
        return self.fc(hidden.squeeze(0))  # [batch_size, output_dim]


model = SentimentRNN(vocab_size=len(vocab), embed_dim=100, hidden_dim=128, output_dim=2).to(DEVICE)
optimizer = optim.Adam(model.parameters(), lr=1e-3)
criterion = nn.CrossEntropyLoss()

# ========== 模型训练 ==========

EPOCHS = 30
train_losses = []

for epoch in range(EPOCHS):
    model.train()
    total_loss = 0
    for texts, labels in train_loader:
        optimizer.zero_grad()
        outputs = model(texts)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    avg_loss = total_loss / len(train_loader)
    train_losses.append(avg_loss)
    print(f"Epoch {epoch + 1}/{EPOCHS}, Loss: {avg_loss:.4f}")

# ========== 简单评估 ==========
model.eval()
correct, total = 0, 0
with torch.no_grad():
    for texts, labels in test_loader:
        outputs = model(texts)
        predicted = outputs.argmax(1)
        correct += (predicted == labels).sum().item()
        total += labels.size(0)
print(f"Test Accuracy: {correct / total:.4f}")

# ========== 可视化训练损失 ==========

plt.plot(range(1, EPOCHS + 1), train_losses, marker='o')
plt.title("Training Loss over Epochs")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.grid(True)
plt.show()

# ========== 导出为 ONNX 模型 ==========

model.eval()
example_input = torch.randint(0, len(vocab), (100, 1)).to(DEVICE)  # [seq_len, batch_size]
torch.onnx.export(
    model, example_input, "../server/models/imdb.onnx",
    input_names=["text"],
    output_names=["output"],
    dynamic_axes={"text": {0: "seq_len", 1: "batch_size"}, "output": {0: "batch_size"}},
    opset_version=11
)

print("✅ 模型已导出为 imdb.onnx")
```

SVD是一个将任何矩阵拆分为三个特定矩阵的数学方法，对于任意矩阵$A(m \times n)$可以用公式表示为：
$$A=US{V^T}$$

其中：
+ $ U $ 是一个 $ m \times m $ 的正交矩阵，包含了矩阵 $ A $ 的左奇异向量。
+ $ S $ 是一个 $ m \times n $ 的对角矩阵，包含了矩阵 $ A $ 的奇异值（非负数），这些奇异值按降序排列。
+ $ V^T $ 是 $ n \times n $ 的正交矩阵，包含了矩阵 $ A $ 的右奇异向量的转置。
