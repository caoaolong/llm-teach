## 正弦-余弦位置编码

$$
PE(pos,2i) = sin(\frac{pos}{10000^{\frac{2i}{d_{model}}}}), PE(pos,2i + 1) = cos(\frac{pos}{10000^{\frac{2i}{d_{model}}}})
$$

+ $PE(pos, i)$表示位置编码矩阵中第$pos$个位置，第$i$个维度的值
+ $d_{model}$表示模型的维度，即词嵌入的向量维度

---

### 使用正余弦函数进行位置编码的决定性因素：

#### 👉1.能够清晰地表达每个词的相对位置关系

选择正弦和余弦函数的原因在于他们的位移性质不变，即**对任意固定偏移量$k$和任意频率$w$，有如下恒等关系：**

$$
\begin{bmatrix}\sin((p+k)w)\\ \cos((p+k)w)\end{bmatrix}
=
\begin{bmatrix}\cos(kw) & \sin(kw)\\ -\sin(kw) & \cos(kw)\end{bmatrix}
\begin{bmatrix}\sin(pw)\\ \cos(pw)\end{bmatrix}
$$

在自然语言处理中，词汇之间的相对位置（例如，“下一个词”、“前两个词”）往往比绝对位置（“第5个词”）更重要。这种线性表达的特性意味着，模型要理解“相距k个位置”这个概念，只需要学习一套固定的线性变换权重。这极大地降低了模型学习相对位置信息的难度，使其变得高效且直观。

#### 👉2.值域有界，提供数值稳定性

无论序列多长（pos值多大），编码值都不会变得极大或极小。这保证了它与词嵌入向量（通常也归一化到较小范围）相加后，不会破坏输入的分布，不会导致梯度爆炸或消失等问题，确保了模型训练的稳定性。


#### 👉3.内在的多频率特性，蕴含丰富的层次化信息

公式中的分母$10000^{\frac{2i}{d_{model}}}$使得维度索引i越小，波长越长（频率越低）；i越大，波长越短（频率越高）。因此，编码向量的不同维度对应了不同的频率。

这种设计让编码包含了丰富的结构化信息。低频维度（变化慢）可以捕捉长程的、粗粒度的位置关系（如句子或段落之间的位置）；高频维度（变化快）可以捕捉短程的、细粒度的位置关系（如相邻词汇的位置）。这为模型提供了多层次的位置感知能力。


#### 👉4.确定性且无需学习，泛化能力极强

正弦余弦编码是一个基于数学公式的确定性计算过程，而非需要通过训练来更新的模型参数。

这带来了强大的泛化能力。模型在训练时可能只见过长度为512的序列，但在推理时完全可以处理长度为1000甚至更长的序列，因为只需要将pos=1000代入公式计算即可。这对于处理长文档至关重要。

```python
import numpy as np
import matplotlib.pyplot as plt

def sinusoidal_position_encoding(position, d_model):
    """计算正弦余弦位置编码"""
    pe = np.zeros(d_model)
    for i in range(0, d_model, 2):
        div_term = np.exp(i * -np.log(10000.0) / d_model)  # 计算分母项
        pe[i] = np.sin(position * div_term)  # 偶数维度用 sin
        pe[i+1] = np.cos(position * div_term)  # 奇数维度用 cos
    return pe

# 设置参数
d_model = 128  # 编码维度（通常与词嵌入维度相同）
max_len = 50   # 最大序列长度

# 生成位置编码矩阵（max_len × d_model）
position_encoding = np.zeros((max_len, d_model))
for pos in range(max_len):
    position_encoding[pos, :] = sinusoidal_position_encoding(pos, d_model)

# 可视化
plt.figure(figsize=(12, 6))
plt.imshow(position_encoding.T, cmap='viridis', aspect='auto')
plt.xlabel('Position (pos)')
plt.ylabel('Embedding Dimension (i)')
plt.colorbar(label='Encoding Value')
plt.title('Sinusoidal Positional Encoding (pos × d_model)')
plt.show()
```

## 旋转位置编码

> 2021年4月中国科学家苏剑林等人发表的论文《RoFormer: Enhanced Transformer with Rotary Position Embedding》中提出了**旋转位置编码（RoPE）**的概念。

RoPE的核心思想非常优雅：它不再将位置信息作为一个独立的“附加项”加到词向量上，而是通过旋转矩阵对词向量本身进行“旋转”，旋转的角度大小取决于该词的位置。

---

假设 embedding 维度是 $d$，我们取每个向量的前两维为一对：

对于一个 token 在位置 $m$，其向量 $\mathbf{x}_m \in \mathbb{R}^d$，可以写作：

$$
\mathbf{x}_m = \big[(x_{m,1}, x_{m,2}), (x_{m,3}, x_{m,4}), \ldots, (x_{m,d-1}, x_{m,d})\big]
$$

RoPE 定义了一个二维旋转矩阵：

$$
R_{\theta_m} =
\begin{bmatrix}
\cos \theta_m & -\sin \theta_m \\
\sin \theta_m & \cos \theta_m
\end{bmatrix}
$$

那么，RoPE 就是对每一对分量应用旋转：

$$
(x_{m,2k-1}, x_{m,2k}) \;\;\mapsto\;\; R_{\theta_{m,k}} \cdot (x_{m,2k-1}, x_{m,2k})^{\top}, \theta_{m,k} = m \cdot \alpha_k, \quad \text{其中 } \alpha_k = 10000^{-2k/d}
$$

也就是说：

* $m$ 表示位置（pos）
* $k$ 表示维度对的索引
* $\alpha_k$ 控制频率，与正弦位置编码一致

对整个向量写作：

$$
\text{RoPE}(\mathbf{x}_m) = \mathbf{x}_m \cdot R_{\Theta_m}
$$

其中 $\Theta_m$ 是由所有维度的 $\theta_{m,k}$ 构成的对角旋转矩阵。


对于 **query 和 key**：

$$
q_m' = \text{RoPE}(q_m), \quad k_n' = \text{RoPE}(k_n)
$$

那么内积变为：

$$
\langle q_m', k_n' \rangle \;=\; \langle q_m, R_{\theta_{m-n}} k_n \rangle
$$

$$
(x_{2k-1}, x_{2k}) \mapsto (x_{2k-1}\cos \theta_{m,k} - x_{2k}\sin \theta_{m,k}, \;\; x_{2k-1}\sin \theta_{m,k} + x_{2k}\cos \theta_{m,k}), \theta_{m,k} = m \cdot 10000^{-\tfrac{2k}{d}}
$$

---

### 解决的问题

1. 传统的正弦编码虽然能计算未见过的位置，但效果会下降。RoPE的旋转机制使其在处理长于训练序列的文本时表现更加优异。

2. 在相对位置计算中，RoPE天然地会给更近的token分配更多的注意力，这与人类的直觉（更关注临近词）相符。

3. RoPE不是简单地将位置信息“加”到$K$和$Q$上，而是将它们“乘”到一个旋转矩阵中，从而在计算Attention Score时自然地融入位置信息。

```python
import numpy as np
import matplotlib.pyplot as plt

def rotate_vector(vec, angle_radians):
    """旋转一个二维向量"""
    rot_matrix = np.array([
        [np.cos(angle_radians), -np.sin(angle_radians)],
        [np.sin(angle_radians),  np.cos(angle_radians)]
    ])
    return np.dot(rot_matrix, vec)

# 假设一个词的嵌入向量在二维空间中（为了可视化）
word_vec = np.array([1.0, 0.5]) # 可以理解为某个注意力头下的q或k向量

# 定义角度：theta_i，这里假设基角度为 15度 (π/12)
base_angle = np.pi / 12
positions = [0, 1, 2, 3] # 四个不同的位置

# 绘制旋转过程
plt.figure(figsize=(10, 8))
colors = ['r', 'g', 'b', 'purple']

for i, pos in enumerate(positions):
    angle = pos * base_angle # 计算旋转角度
    rotated_vec = rotate_vector(word_vec, angle)

    # 绘制原始向量
    if i == 0:
        plt.quiver(0, 0, word_vec[0], word_vec[1], angles='xy', scale_units='xy', scale=1, color='black', label='Original Vector')
    # 绘制旋转后的向量
    plt.quiver(0, 0, rotated_vec[0], rotated_vec[1], angles='xy', scale_units='xy', scale=1, color=colors[i], label=f'Pos {pos} (θ={np.degrees(angle):.1f}°)')

plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.axhline(y=0, color='k', linestyle='--', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='--', alpha=0.3)
plt.grid(True)
plt.legend()
plt.title('RoPE: Rotating a Word Vector Based on Its Position')
plt.show()
```
